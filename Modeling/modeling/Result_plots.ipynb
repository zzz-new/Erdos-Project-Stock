```python
apple_train=pd.read_csv('../../Results/Apple_train.csv')
apple_train.head(3)
```

```python
Apple_test_ARIMA = pd.read_csv('../../Results/Apple_test_ARIMA.csv')
Apple_test_ARIMA.head(3)
```

```python
Apple_test_ARIMA["Date_test"] = pd.to_datetime(Apple_test_ARIMA["Date_test"])
```

```python
Apple_test_ARIMA_dyn_3features = pd.read_csv('../../Results/Apple_test_ARIMA_dyn_3features.csv')
Apple_test_ARIMA_dyn_3features["Date_test"] = pd.to_datetime(Apple_test_ARIMA_dyn_3features["Date_test"])
Apple_test_ARIMA_dyn_3features.head(3)
```

```python
Apple_test_ARIMA_dyn_8features = pd.read_csv('../../Results/Apple_test_ARIMA_dyn_8features.csv')
Apple_test_ARIMA_dyn_8features["Date_test"] = pd.to_datetime(Apple_test_ARIMA_dyn_8features["Date_test"])
Apple_test_ARIMA_dyn_8features.head(3)
```

```python
Apple_test_LSTM = pd.read_csv('../../Results/df_test_LSTM.csv')
Apple_test_LSTM["Date_test"] = pd.to_datetime(Apple_test_LSTM["Date_test"])
Apple_test_LSTM.head(3)
```

```python
Apple_test_LSTM_add = pd.read_csv('../../Results/df_test_LSTM_add.csv')
Apple_test_LSTM_add["Date_test"] = pd.to_datetime(Apple_test_LSTM_add["Date_test"])
Apple_test_LSTM_add.head(3)
```

```python
plt.figure(figsize=(20,10))


plt.plot(apple_train.Date[:], apple_train.Price_train[:], 'bo', label="Training Data")

plt.plot(Apple_test_ARIMA.Date_test[:], Apple_test_ARIMA.Price_test[:], 'r', label="Test Data")

plt.plot(Apple_test_ARIMA.Date_test[:], Apple_test_ARIMA.Forecast_ARIMA[:], 'k', label="ARIMA (Baseline)")


plt.xlabel("Date", fontsize=18)
plt.ylabel("Apple Price", fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=2)
plt.title('Apple Price Forecast', fontsize = 24)

plt.show()
```
