```python
apple_raw=pd.read_csv('../../Dataset/AAPL.csv')
```

```python
apple_raw["Date"] = pd.to_datetime(apple_raw["Date"])
```

```python
apple_raw
```

### Train-test split

```python
train_split_ratio=0.75
```

```python
# split dataset into train and test
len_total=apple_raw.shape[0]
len_train=int(len_total*train_split_ratio)
len_test = len_total - len_train
print(f"len_total:{len_total}", f"len_train:{len_train}", f"len_test:{len_total - len_train}")
```

```python
feature_list = ['Close']
```

```python
# pick the "Close" column
# need to reshape as (len, 1)
apple_raw_all = np.array(apple_raw[feature_list].iloc[:])
apple_raw_train = np.array(apple_raw[feature_list].iloc[:len_train])
apple_raw_test = np.array(apple_raw[feature_list].iloc[len_train:])
```

```python
[np.shape(apple_raw_train), np.shape(apple_raw_test)]
```

### Scaling and transformation

```python
# scale = MinMaxScaler(feature_range = (0,1))
# apple_raw_train_scaled = scale.fit_transform(apple_raw_train)
# apple_raw_test_scaled = scale.fit_transform(apple_raw_test)
```

```python
scale = MinMaxScaler(feature_range = (0,1))

apple_raw_scaled = scale.fit_transform(apple_raw_all)
apple_raw_train_scaled = apple_raw_scaled[:len_train]
apple_raw_test_scaled = apple_raw_scaled[len_train:]
```

```python
apple_raw_test_scaled.shape
```

```python
# set window, use previous __ days as input to calculate the next day output
window = 60
```

```python
x_train = []
y_train = []

for i in range(window, len_train):
    x_train.append(apple_raw_train_scaled[i-window:i, 0])  # every x_train[i] is an array of len window=60
    y_train.append(apple_raw_train_scaled[i,0])            # every y_train[i] is a number of len 1
    
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
# reshape for LSTM input as (len of training set,  len of each input, len of each output)
# each timestep is a feature tensor
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train.shape
```

```python
y_train
```

```python
# do the same for x_test
x_test = []

for i in range(window, len_test):
    x_test.append(apple_raw_test_scaled[i-window:i, 0])
    
x_test = np.array(x_test)
```

```python
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
[x_test.shape, len_test]
```

## build and compile RNN

```python
keras.backend.clear_session()
```

```python
rnn = Sequential()
```

```python
rnn.add(LSTM(units = 128, return_sequences = True, input_shape = (x_train.shape[1],1)))
rnn.add(Dropout(0.2))
```

```python
rnn.add(LSTM(units = 64, return_sequences = True))
rnn.add(Dropout(0.2))
```

```python
rnn.add(LSTM(units = 50,return_sequences = False))
rnn.add(Dropout(0.2))
```

```python
#rnn.add(LSTM(units = 64))
#rnn.add(Dropout(0.2))
```

```python
rnn.add(Dense(units = 1))
```

```python
rnn.summary()
```

```python
rnn.compile(optimizer = 'adam', loss = 'mse', metrics=['accuracy'])
```

```python
# history = 
rnn.fit(x_train,y_train,epochs = 25, batch_size = 32)
```

```python
rnn.save('apple_rnn_models/lstm_scale_all_128_64_50_1')
```

```python
LSTM_predict_scale = rnn.predict(x_test)
```

```python
# initialize pred_window
pred_window = np.array([x_test[0]])
forecast_result = np.array([])
#len_window = 60


# forecast the future price
for i in range(len(x_test)):
    print(i,'/',len(x_test))
    forecast_price = rnn.predict(pred_window)
    
    forecast_result = np.append(forecast_result, forecast_price)

    # update pred_window
    pred_window = np.append(pred_window[0][1:], forecast_price).reshape(1,window,1)
```
