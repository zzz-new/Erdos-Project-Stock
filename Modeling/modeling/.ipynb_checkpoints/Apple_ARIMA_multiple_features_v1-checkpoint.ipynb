```python
apple_raw=pd.read_csv('../../Dataset/merge_data.csv')
apple_raw.head(3)
```

```python
apple_raw=apple_raw.rename(columns={"Date_x": 'Date'})
```

```python
apple_raw["Date"] = pd.to_datetime(apple_raw["Date"])
```

```python
apple_raw=apple_raw.drop(columns=['grouper','Date_y','Unnamed: 0'])
```

```python
plt.figure(figsize=(20,8))

plt.plot(apple_raw.Date[-200:], 
         apple_raw.Close[-200:],
         'b',
         label="Training Data")

plt.xlabel("Date", fontsize=18)
plt.ylabel("Closing Price", fontsize=18)
plt.title('APPLE', fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=2)

plt.show()


plt.figure(figsize=(20,8))

plt.plot(apple_raw.Date[:], 
         apple_raw.Close[:],
         'b',
         label="Training Data")

plt.xlabel("Date", fontsize=18)
plt.ylabel("Closing Price", fontsize=18)
plt.title('APPLE', fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=2)

plt.show()
```

```python
apple_train = apple_raw.iloc[:-500].copy()
apple_test = apple_raw.iloc[-500:].copy()
```

```python
[apple_train.shape, apple_test.shape, len(apple_raw)]
```

```python
apple_train
```

```python
# ARIMA Model(order=(p,d,q))
model_arima_dyn = ARIMA(endog=apple_train.Close.values, 
                        exog= apple_train[['GDP_growth']], 
                        order=(5,1,3)).fit()
```

```python
print(model_arima_dyn.summary())
```

```python
plt.figure(figsize=(20,10))


plt.plot(apple_train.Date[-800:], 
         apple_train.Close[-800:],
         'b',
         label="Training Data")


## We can get the fitted values with fittedvalues
plt.plot(apple_train.Date[-800:], 
         model_arima_dyn.fittedvalues[-800:],
         'g--',
         label="ARIMA(5,1,3) Fit")

plt.plot(apple_test.Date[:], 
         apple_test.Close[:],
         'r',
         label="Test Data")

## We can get the forecast with .forecast(h)
plt.plot(apple_test.Date[:], 
         model_arima_dyn.forecast(len(apple_test), exog = apple_test[['GDP_growth']][:]),
         'r--o',
         label="ARIMA(5,1,3) Forecast")

plt.xlabel("Date", fontsize=18)
plt.ylabel("Index", fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=2)

plt.show()
```

```python
model_arima_dyn.forecast(len(apple_test), exog = apple_test[['GDP_growth']][:])
```

```python
model_arima_dyn.predict(len(apple_test))
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true))
```

```python
mape_arima_dyn = mean_absolute_percentage_error(apple_train[['Close']].values, 
                                                model_arima_dyn.predict(len(apple_test)))
mape_arima_dyn
```
