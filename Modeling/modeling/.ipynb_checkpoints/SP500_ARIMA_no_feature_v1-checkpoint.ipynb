```python
sp500index=pd.read_csv('../../Dataset/sp500_index.csv')
```

```python
sp500index
```

```python
sp500index['Date']
```

```python
sp500index['Date_new'] = sp500index['Date'].astype('datetime64[ns]')
```

```python
sp500=sp500index.rename(columns={"S&P500": "sp500"}).copy(deep=True)
```

```python
plt.figure(figsize=(20,8))

plt.plot(sp500.Date_new[-0:], 
         sp500.sp500[-0:],
         'b',
         label="Training Data")

plt.xlabel("Date", fontsize=18)
plt.ylabel("Closing Price", fontsize=18)
plt.title('S&P500', fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=3)

plt.show()
```

```python
sp500_train = sp500.iloc[:-500].copy()
sp500_test = sp500.iloc[-500:].copy()
```

```python
[sp500_train.shape,sp500_test.shape]
```

```python
sp500_train
```

```python
# ARIMA Model(order=(p,d,q))
model_arima = ARIMA(sp500_train.sp500, order=(5,1,3)).fit()
```

```python
print(model_arima.summary())
```

```python
plt.figure(figsize=(20,10))


plt.plot(sp500_train.Date_new[-400:], 
         sp500_train.sp500[-400:],
         'b',
         label="Training Data")


## We can get the fitted values with fittedvalues
plt.plot(sp500_train.Date_new[-400:], 
         model_arima.fittedvalues[-400:],
         'g--',
         label="ARIMA(5,1,3) Fit")

plt.plot(sp500_test.Date_new[:50], 
         sp500_test.sp500[:50],
         'r',
         label="Test Data")

## We can get the forecast with .forecast(h)
plt.plot(sp500_test.Date_new[:50], 
         model_arima.forecast(50),
         'r--o',
         label="ARIMA(5,1,3) Forecast")

plt.xlabel("Date", fontsize=18)
plt.ylabel("Index", fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14, loc=2)

plt.show()
```
