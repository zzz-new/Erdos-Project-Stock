### Train-test split

```python
len_total=apple_raw.shape[0]
len_train=int(len_total*0.75)
len_test = len_total - len_train
print(f"len_total:{len_total}", f"len_train:{len_train}", f"len_test:{len_total - len_train}")
```

```python
# pick the "Close" column
apple_raw_train = apple_raw.iloc[:len_train, 4:5].values
apple_raw_test = apple_raw.iloc[len_train:, 4:5].values
```

```python
[np.shape(apple_raw_train), np.shape(apple_raw_test)]
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler(feature_range = (0,1))
apple_raw_train_scaled = scale.fit_transform(apple_raw_train)
apple_raw_test_scaled = scale.fit_transform(apple_raw_test)
```

```python
apple_raw_train_scaled
```

```python
apple_raw_test_scaled.shape
```

```python
# Create a data structure with 60-time steps and 1 output

x_train = []
y_train = []

for i in range(60,len_train):
    x_train.append(apple_raw_train_scaled[i-60:i, 0])
    y_train.append(apple_raw_train_scaled[i,0])
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_train[1]
```

```python
y_train
```

```python
y_train[0]
```

```python
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_test = []

for i in range(60,len_test):
    x_test.append(apple_raw_test_scaled[i-60:i,0])
```

```python
x_test = np.array(x_test)
```

```python
[x_test.shape, len_test]
```

```python
x_test[0]
```

```python
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
x_test.shape
```

```python
x_test[0]
```

## build and compile RNN

```python
rnn = Sequential()
```

```python
rnn.add(LSTM(units = 50, return_sequences = True, input_shape = (x_train.shape[1],1)))
rnn.add(Dropout(0.2))
```

```python
rnn.add(LSTM(units = 50, return_sequences = True))
rnn.add(Dropout(0.2))
```

```python
rnn.add(LSTM(units = 50,return_sequences = False))
rnn.add(Dropout(0.2))
```

```python
#rnn.add(LSTM(units = 50))
#rnn.add(Dropout(0.2))
```

```python
rnn.add(Dense(units = 1))
```

```python
rnn.summary()
```

```python
rnn.compile(optimizer = 'adam', loss = 'mse', metrics=['accuracy'])
```

```python
validation_split=0.1
```

```python
# history = 
rnn.fit(x_train,y_train,epochs = 50, batch_size = 32)
```

```python
rnn.save('apple_rnn_models/lstm')
```

```python
[x_train.shape, x_test.shape]
```

```python
LSTM_predict_scale = rnn.predict(x_test)
```

```python
LSTM_predict = scale.inverse_transform(LSTM_predict_scale)
```
